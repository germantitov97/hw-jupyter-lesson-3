# learning and exercises

```python
a = np.array([1, 2, 3])
b = np.array([1, 5, 3])
print(a == b)
print(a != b)

print(a[a == b])
```

```python
a = np.array([1, 2, 3])
print(a < 2)
print(a >= 2)

vals = np.array([5, 12, 3, 8, 1])
print(vals[vals > 5])
```

```python
print(np.array_equal(np.array([1, 2]), np.array([1,2])))
print(np.array_equal(np.array([1,2]), np.array([1,9])))
print(np.array_equal(np.array([1,2]), np.array([[1,2]])))
np.allclose(np.array([0.1+0.2]), np.array([0.3]))
```

```python
a = np.array([1, 2, 3, 4])
print(a.sum())

b = np.array([[1,2,3],[4,5,6]])
print(b.sum(axis=0))
print(b.sum(axis=1))
print(b.sum())
```

```python
b = np.array([[1,2,3],
              [4,2,6]])
print(b.min())
print(b.max(axis=0))
print(b.min(axis=1))
print(b.argmax())
```

```python
a = np.array([4,1,9,2,7])
print(a.argmin())
print(a.argmax())

b = np.array([[1,5,3],
              [4,2,6]])
print(b.argmax(axis=0))
print(b.argmax(axis=1))
```

```python
a = np.arange(0,10)
print(a.mean())

b = np.array([[1,2,3],
              [4,5,6]])
print(b.mean(axis=0))
print(b.mean(axis=1))
```

```python
a = np.array([1,3,2,5,4])
print(np.median(a))

b = np.array([1,3,2,5,4,1000])
print(np.median(b))
print(b.mean())
```

```python
x = np.array([1,2,3])
y = np.array([4,5,6])
print(np.corrcoef(x,y))

r = np.corrcoef(x,y)[0,1]
print(r)
```

```python
a = np.array([2,4,4,4,5,5,7,9])
print(np.std(a))
print(np.std(a, ddof=1))
print(a.mean())
```

```python
a = np.array([2,4,4,4,5,5,7,9])
print(np.var(a))
print(np.std(a)**2)
print(np.var(a, ddof=1))
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
print(np.concatenate((a,b)))

m1 = np.array([[1,2],[3,4]])
m2 = np.array([[5,6],[7,8]])
print(np.concatenate((m1,m2),axis=0))

print(np.concatenate((m1,m2),axis=1))
```
